First, we load the data.

```python
# load data
ratings = pd.read_csv('ratings.csv')

# sample dataset
# be careful, large dataset!
ratings = ratings[:10000]

print(ratings.head())

# print some information
noMovies = len(ratings['movieId'].unique())
noUsers = len(ratings['userId'].unique())
print(str(noMovies)+" from "+str(noUsers)+' users')
```

Create an empty utility matrix

```python
utility = np.zeros(shape=(noUsers,noMovies))
```

Store movieIds as indices to use in utility matrix as the current indices don't match the sequential indices that a matrix uses.

```python
movieIds = {}
midi = 0
for value in ratings['movieId'].unique():
    movieIds[value]=midi
    midi = midi + 1
```

Populate the utility matrix by looping all the rows in the ratings dataframe

```python
for index, line in ratings.iterrows():
    uid = int(line['userId'])-1
    mid = movieIds[line['movieId']]
    rating = line['rating']
    # store the rating in the utility matrix at row user id - uid and column movie - mid
    utility[uid,mid]=rating
```
